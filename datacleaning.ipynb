```python
def find_gender(x):
    if x=='Male':
        return 1
    else:
        return 0
data['GENDER']=data['GENDER'].apply(find_gender)
```

```python
data.head()
```

```python
data['LIVWTH'].value_counts()
```

```python
def find_education(x):
    if x=='Graduate':
        return 0
    elif x=='HSC':
        return 1
    elif x=='Post Graduate':
        return 3
    else:
        return 4

data['EDU']=data['EDU'].apply(find_education)
```

```python
data['PROF'].value_counts()
```

```python
def find_profession(x):
    if x=='Service holder':
        return 0
    elif x=='Unemployed':
        return 1
    elif x=='Other':
        return 3
    elif x=='Service holder':
        return 4
    else:
        return 5

```

```python
def maritial(x):
    if x=='Unmarried':
        return 0
    else:
        return 1
    
```

```python
def living_with(x):
    if x=='With Family':
        return 1
    else:
        return 0
```

```python
def binary(x):
    if x=='Yes':
        return 1
    else:
        return 0
```

```python
data.head()
```

```python
data['PROF']=data['PROF'].apply(find_profession)
```

```python
data['MARSTS']=data['MARSTS'].apply(maritial)
```

```python
data['LIVWTH']=data['LIVWTH'].apply(living_with)
```

```python
data['ENVSAT']=data['ENVSAT'].apply(binary)
```

```python
data['POSSAT']=data['POSSAT'].apply(binary)
data['FINSTR']=data['FINSTR'].apply(binary)
```

```python
def find_age(x):
    if x=='21-25':
        return 1
    elif x=='26-30':
        return 2
    elif x=='16-20':
        return 3
    elif x=='46-50':
        return 4
    elif x=='51-55':
        return 5
    elif x=='31-35':
        return 6
    elif x=='41-45':
        return 7
    elif x=='36-40':
        return 8
    elif x=='60+':
        return 9
    else:
        return 10
```

```python
data['AGERNG']=data['AGERNG'].apply(find_age)
```

```python
data['RESDPL'].value_counts()
```

```python
def find_destination(x):
    if x=='City':
        return 1
    elif x=='Town':
        return 2
    else:
        return 3
        
```

```python
data['RESDPL']=data['RESDPL'].apply(find_destination)
```
